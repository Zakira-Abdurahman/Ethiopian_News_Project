```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
url = "https://www.waltainfo.com/"
```

```python
# send request
response = requests.get(
    url,
    headers={"User-Agent": "Mozilla/5.0"}
)
print("Status Code:", response.status_code)
```

```python
#parse html
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.title)
print(soup.title.text)
```

```python
print("Articles:", len(soup.find_all("article")))
print("H1:",len(soup.find_all("h")))
print("H2:", len(soup.find_all("h2")))
print("H3:", len(soup.find_all("h3")))
print("Div:",len(soup.find_all("div")))
print("Links:" ,len(soup.find_all("a")))
```
