```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
url = "https://www.waltainfo.com/"
```

```python
# send request
response = requests.get(
    url,
    headers={"User-Agent": "Mozilla/5.0"}
)
print("Status Code:", response.status_code)
```

```python
#parse html
soup = BeautifulSoup(response.text, "html.parser")
```

```python
#confirm if it is Correct Page
print(soup.title)
print(soup.title.text)
```

```python
#investigate the html
print("Articles:", len(soup.find_all("article")))
print("H1:",len(soup.find_all("h")))
print("H2:", len(soup.find_all("h2")))
print("H3:", len(soup.find_all("h3")))
print("Div:",len(soup.find_all("div")))
print("Links:" ,len(soup.find_all("a")))
```

```python
#print first 3000 characters
print(soup.prettify()[:3000])
```

```python
# check what h2 tags contain, does h2 contain the title, link and news url
h2_tags = soup.find_all("h2")
print("Total H2:", len(h2_tags))
for i, tag in enumerate(h2_tags[:5], start=1):
    print(f"\nH2 {i}")
    print(tag)
```

```python
#check the first article in h2, print title and url
first_news = soup.find("h2", class_="entry-title")
print(first_news)

title = first_news.get_text(strip=True)
print(title)

link = first_news.find("a")["href"]
print(link)

```

```python
#find all news title in h2 with entry-title
news_items = soup.find_all("h2", class_="entry-title")
print("Total News Found:", len(news_items))

#extract title and url for every araticle
data = []
for news in news_items:
    a_tag = news.find("a")
    title = a_tag.get_text(strip=True)
    url= a_tag.get("href")
    data.append({
        "title": title,
        "url": url
    })
print("Collected", len(data))

#verify the results
df = pd.DataFrame(data)
df.head()
```

```python
#find what parent container contain
print(news_items[0].parent.prettify())
```

```python
#improve scraper to contain title, url, date,author and catagories
data = []
for news in news_items:
    #title and url
    a_tags = news.find("a")
    title = a_tags.get_text(strip=True)
    url = a_tags.get("href")

    #parent container
    container = news.parent

    #date
    date = container.find("div", class_="date")
    date = date.get_text(strip=True) if date else ""

    #Author
    author = container.find("div", class_="by-author")
    author = author.get_text(strip=True) if author else ""

    #categories
    cat_links = container.find("div", class_="cat-links")
    if cat_links:
        categories = ",".join(
            a.get_text(strip=True)
            for a in cat_links.find_all("a")
        )
    else:
        categories = ""
        
    data.append({
        "title": title,
        "url": url,
        "date": date,
        "author": author,
        "categories": categories
    })

df = pd.DataFrame(data)
df.head()

#check dataframe info
df.info()
df.isnull().sum()

#check for duplication 
print("Duplicate rows:", df.duplicated().sum())
print(df[["title", "date"]])

df = df.drop_duplicates(subset=["url"])

#reset index
df = df.reset_index(drop=True)
print("Final number of articles:", len(df))
    

#save to csv
df.to_csv(
    "waltainfo_news.csv",
    index=False,
    encoding="utf-8"
)
print("CSV file saved successfully")

#save to json
with open("waltainfo_news.json", "w", encoding="utf-8") as file:
    json.dump(
        df.to_dict(orient="records"),
        file,
        ensure_ascii=False,
        indent=4
    )
print("JSON file saved successfully")

              
```
